```python
# Convert time duration strings to seconds
def time_to_seconds(time_str):
    if isinstance(time_str, str) and ':' in time_str: # Check if it's a string AND contains ':'
        hours, minutes, seconds = map(int, time_str.split(':'))
        return hours * 3600 + minutes * 60 + seconds
    return time_str  # Handle cases where the value is already numeric or not in the expected format

raw_ob_data['AHTCalc'] = raw_ob_data['AHTCalc'].apply(time_to_seconds)
raw_ob_data['AHTCalc'] = raw_ob_data['AHTCalc'].astype(float)
raw_ob_data['ACWTime'] = raw_ob_data['ACWTime'].apply(time_to_seconds)
raw_ob_data['ACWTime'] = raw_ob_data['ACWTime'].astype(float)
raw_ob_data['TalkTime'] = raw_ob_data['TalkTime'].apply(time_to_seconds)
raw_ob_data['TalkTime'] = raw_ob_data['TalkTime'].astype(float)
raw_ob_data['ConsultTime'] = raw_ob_data['ConsultTime'].apply(time_to_seconds)
raw_ob_data['ConsultTime'] = raw_ob_data['ConsultTime'].astype(float)
print(raw_ob_data.dtypes)
```

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical columns
numerical_columns = raw_ob_data.select_dtypes(include=["int64", "float64"]).columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 2, idx)
    sns.histplot(raw_ob_data[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(raw_ob_data[feature].skew(), 2)}")

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

```python
#Pearson correlations matrix
corr_ob = raw_ob_data.select_dtypes(include=['float64', 'int64']).corr(method = 'pearson') # Select only numerical columns for correlation calculation
corr_ob
```

```python
# Correlations between Transferred Calls and Hold Calls
sub_corr = np.corrcoef(corr_ob['Transferred'],corr_ob['HoldCalls'])
print('Correlations between Transfer and Hold calls\n',sub_corr)
```

```python
plt.figure(figsize=(8,10), dpi =500)
sns.heatmap(corr_ob,annot=True,fmt=".2f", linewidth=.5)
plt.show()
```
