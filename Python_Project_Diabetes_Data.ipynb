```python
#Median Impute
for i in ["num_procedures","number_outpatient","number_emergency","number_inpatient"]:
    df[i].replace({0:df[i].median()},inplace = True)
```

```python
for col in df.columns:
  if df[col].dtype == object:
    print(col, df[col][df[col] == '?'].count())
```

```python
print ('gender', df[ 'gender'][df[ 'gender'] == 'Unknown/Invalid'].count())
```

```python
df = df.drop(['weight', 'payer_code', 'medical_specialty'], axis = 1)
```

```python
df['race'] = df['race'].replace('?','others')
```

```python
drop_Idx = set(df[(df['diag_1'] == '?') & (df['diag_2'] == '?') & (df['diag_3'] == '?')].index)
```

```python
drop_Idx = set(df[(df['diag_1'] == '?') & (df['diag_2'] == '?') & (df['diag_3'] == '?')].index)
drop_Idx = drop_Idx.union (set(df['gender'][df['gender'] == 'Unknown/Invalid'].index))
drop_Idx = drop_Idx.union(set(df[df['discharge_disposition_id'] == 11].index))
drop_Idx = drop_Idx.union(set(df[df['discharge_disposition_id'] == 19].index))
drop_Idx = drop_Idx.union(set(df[df['discharge_disposition_id'] == 20].index))
drop_Idx = drop_Idx.union(set(df[df['discharge_disposition_id'] == 21].index))
new_Idx = list(set(df.index) - set(drop_Idx))
df = df. iloc [new_Idx]
```

```python
# Handle non-numeric columns before calculating correlations
corrmat = df.select_dtypes(include=['number']).corr()  # Select only numeric columns
plt.subplots(figsize=(10, 10))
sns.heatmap(corrmat, annot=True, square=True)
```

```python
df_dbmed = df.groupby('diabetesMed')['gender'].value_counts()

sns.barplot(x=df_dbmed.index.get_level_values(0),
            y=df_dbmed.values,
            hue=df_dbmed.index.get_level_values(1))
plt.xlabel("Diabetes Medicine")
plt.ylabel("Count")
plt.title("Distribution of Gender by Diabetes Medicine")
plt.show()
```

**Based on the graph above**, Females are the dominant gender taking Diabetes Medication in comparison with Males.

```python
df_raceReadmitted = df.groupby(['race', 'readmitted'])['encounter_id'].count().reset_index(name='count') # count encounter_id and reset index

sns.barplot(data=df_raceReadmitted, x="readmitted", y="count", hue="race")

plt.xlabel("Readmitted")
plt.ylabel("Count")
plt.title("Distribution of Readmission by Race")
plt.show()
```

Across all three value types of No Readmission, Readmitted in less than 30 days, and Readmitted in greater than 30 days, **Caucasians** tend to be the race with the highest prevalence of readmissions with the second highest race being **African Americans.**

```python
age_dict = {'[0-10)':5, '[10-20)':15, '[20-30)':25, '[30-40)':35, '[40-50)':45, '[50-60)':55, '[60-70)':65, '[70-80)':75, '[80-90)':85, '[90-100)':95}
df['age'] = df.age.map(age_dict)
df['age'] = df['age'].astype(int)
df['age'].value_counts()
```

```python
df_ageMaxGlucose = df.groupby(['age', 'max_glu_serum'])['encounter_id'].sum().reset_index(name='count')

sns.set_theme(style="darkgrid")

# use the scatterplot function
sns.scatterplot(
    data=df_ageMaxGlucose,
    x="age",
    y="count", # Update y to count the occurrences
    hue = 'max_glu_serum', # Use max_glu_serum as hue
    alpha=0.7,
    s=100
)

# show the graph
plt.show()
```

Between ages 45-85 we see the highest counts for individuals with a Glucose Serum Level greater than 300 units.
